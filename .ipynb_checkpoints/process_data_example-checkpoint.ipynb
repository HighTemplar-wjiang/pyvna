```python
%matplotlib inline
import re
import os
import csv
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['savefig.dpi'] = 300
```

```python
def _s2y(s):
    """Convert S-parameters to Z-parameters

    Formula: :math: `z = \frac{1+s}{1-s} \times 50`

    Args:
        s (complex): S-parameter to be converted

    Returns:
        complex: converted Z-parameter
    """
    y = ((1.-s) / (1.+s)) / 50.
    return y
```

```python
datapath = "./data/"
re_name = re.compile("S11_(\d{2})mm.csv")
file_list = os.listdir(datapath)

all_labels = []
all_data = []
for filename in file_list:
    # Check if filename matches the patter.
    re_result = re_name.search(filename)
    if re_result:
        
        label = int(re_result.group(1))
        data = []
        print(filename)
        with open(datapath+filename, "r") as f:
            reader = csv.reader(f, delimiter=",")
            head = next(reader)
            for row in reader:
                data.append([float(item) for item in row])
        all_labels.append(label)
        all_data.append(data)

all_data = np.array(all_data)
print(all_data.shape)
```

```python
s11_all = []
y11_all = []
for data in all_data:
    freq_list = data[:, 0]
    s11 = data[:, 1] + 1j * data[:, 2]
    y11 = _s2y(s11)
    s11_all.append(s11)
    y11_all.append(y11)
s11_all = np.array(s11_all)
y11_all = np.array(y11_all)

fig, ax1 = plt.subplots()
plt.xlabel("Frequency (MHz)")
plt.ylabel("Real part of S11 (S)")
for s11 in s11_all:
    max_idx = np.argmax(s11.real)
    # resonant_freq_list.append(freq_list[max_idx])
    #Q_list.append(calculate_Q(freq_list, y11))
    plt.plot(freq_list / 1e6, s11.real)
plt.legend(["d=" + str(label) + " mm" for label in all_labels])
```
